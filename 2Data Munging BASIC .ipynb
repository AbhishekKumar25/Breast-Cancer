```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
```

```python
missing = np.nan
series_obj =Series(['row 1', 'row 2', missing, 'row 5','row 6', missing, 'row 7'])
series_obj
```

#Using objectname.isnull() method to find the missing values.

```python
series_obj.isnull()
```

Filling in the missing values

```python
np.random.seed(25)
DF_obj =DataFrame(np.random.rand(36).reshape(6,6)) #remeber DataFrame is a constructor in this defimned dataobject
DF_obj
```

making few values missin in this randomly generated data

```python
#DataObjectname.loc[row from 3 to 5 and colum 0]
DF_obj.loc[3:5,0] =missing
DF_obj.loc[1:4, 5] = missing
DF_obj
```

Now lets fill the missing values in this object by using 
objectname.fillna(numeric value) function

```python
filled_DF=DF_obj.fillna(0)
filled_DF
```

we can pass a dict in the arg for this fillna function to fill specific values with respect to there place
for example in this example replace 0 with 0.1 NS 5 WITH 1.25 in the dataobject

```python
filled_DF = DF_obj.fillna({0:0.1, 5: 1.25})
filled_DF
```

```python
#fillforward methos ffill repeating the last value to fill the other misasing valyes in the data object.
filled_DF = DF_obj.fillna(method='ffill')
filled_DF
```
